#### Outliers detection

There are some techniques to handle outliers like transformation to square root, logarithmic, boxcox, or winsorization capping. but in the term of data analyzing, those techniques wouldnt be appropriate because it can alter the insights of the data itself. Therefor, our choices are limited to removing or leaving it as they are. We will detect it based on profit because we want to be focused on analyzing the profits made by the products.

```python
#finding outliers by inter quartile range method

def outlierbounds(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    ub = q3 + 1.5*iqr
    lb = q1 - 1.5*iqr
    return ub, lb

#setting outlierdf and inrange df

outdf = pd.DataFrame()
indf = pd.DataFrame()

#detecting outliers based on products

for i in df.Product.unique():
    dfx = df[df.Product == i].copy()
    ub,lb = outlierbounds(dfx.Profit)
    
    outliers = dfx[(dfx.Profit < lb) | (dfx.Profit > ub)]
    inrange = dfx[(dfx.Profit >= lb) & (dfx.Profit <= ub)]
    outdf = pd.concat([outdf,outliers], axis = 0)
    indf = pd.concat([indf, inrange], axis = 0)
    
pc = round(len(outdf) / len(df), 3) * 100
print(f'Outliers count and percentage based on profits per product: {len(outdf)} ({pc}%)')
```

There are ~16% of whole rows that are considered as outliers. This number is quite big and surely we can extract more insights if we conduct holistic analysis to these rows. These can be affected by factors like seasonal trend, market trend, or unexpected marketing performances. There is also possibilities that these rows are the result of wrongly inputted data. These rows can be used for further analysis and can be used as a comparison to the majority of the data. 

***But from now on, we will be focused on the data that is in the range of the product profits majority. Again, I understand that the data that considered as outliers may carry valuable informations. But for the sake of simplicity and avoiding too much information, we will be focused by majority of the data only***

***Reasoning behind detecting outliers per product profit and not as a whole:***

The profits and sales generated per product are difference. And the gap is large that the smaller profit or sales products will treat the higher profit or sales products as outliers. This is not right because we want to see the performances of the products as themselves.
```python
import scipy.stats as stats

emplist = []
for i in list(df['Product'].unique()):
    x = list(df[df['Product'] == i].Sales)
    emplist.append(x)
    

# Perform Kruskal-Wallis test
h_statistic, p_value = stats.kruskal(*emplist)

# Print results
print("H-statistic:", h_statistic)
print("P-value:", p_value)

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is significant difference in medians.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in medians.")
```

we can see that there are clear gaps between sales and profits per product. This may be affected by the prices of the product itself or other unknown factors.

```python
#conducting normality test with shapiro-wilk

from scipy.stats import shapiro
alpha = 0.05

for i,ii in enumerate(df.Product.unique()):
    checkdf = indf[indf['Product']==ii].Profit

    _,pvalshap = shapiro(checkdf)
    
    print(ii, 'Product Profit Distribution')
    
    if pvalshap > alpha:
        print("Pval: ",round(pvalshap, 2),"\nFail to reject the null hypothesis:", 
              "Sample data comes from a normally distributed population.")
    else:
        print("Pval: ",round(pvalshap, 2), "\nReject the null hypothesis: Sample", 
              "data does not come from a normally distributed population.\n")
```

We can see that, even though some of the distributions are closely bell shaped, all of the product profit distributions are not normally distributed. the pval closer to our tolerance value is on marketing suite product with 0.04. we can also see from the histogram, the distribution for this product is the one that is most closely bell shaped.

```python
#feature engineering steps

#change order date to datetime and creating time fragments
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date MY'] = df['Order Date'].dt.strftime('%Y-%m')
df['Order Date Y'] = df['Order Date'].dt.strftime('%Y')
df['Order Date M'] = df['Order Date'].dt.strftime('%m')


#make discount amount given in total
mask = df['Discount'] > 0
df.loc[mask, 'DiscountGiven'] = (df.loc[mask, 'Sales'] / (1 - df.loc[mask, 'Discount'])) * df.loc[mask, 'Discount']
df['DiscountGiven'] = df['DiscountGiven'].fillna(0)

df['ProfitPercentageSales'] = round((df['Profit'] / df['Sales']) * 100, 2)

#make expenses column that defines Cost of Goods Sold (expenses)
df['Expenses'] = df['Sales'] + df['DiscountGiven'] - df['Profit']
df['ProfitPercentageExpenses'] = round((df['Profit'] / df['Expenses']) * 100, 2)
df.head()
```

We add new columns to the data. They are:

- Order Date MY : Fragment of order date that get the month and year of transaction
- Order Date Y : Fragment of order date that get the year of transaction
- Order Date M : Fragment of order date that get the month of transaction
- DiscountGiven : Amount of discount given in transaction
- Expenses : Cost of Goods Sold (COGS) in one transaction
- ProfitPercentageSales : Percentage of profit against the sales amount made
